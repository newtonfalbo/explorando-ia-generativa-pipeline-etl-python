# Explorando IA Generativa em um Pipeline de ETL com Python



**Cenário**

Uma instituição financeira deseja utilizar Inteligência Artificial Generativa para enriquecer dados de clientes e oferecer recomendações financeiras personalizadas.

Neste projeto, foi desenvolvido um pipeline ETL (Extract, Transform and Load) em Python capaz de:

Extrair dados de clientes a partir de uma fonte estruturada;
Realizar transformações e tratamentos nos dados;
Utilizar IA Generativa para criar insights personalizados;
Armazenar os resultados enriquecidos para posterior análise.

O objetivo é demonstrar como técnicas de Engenharia de Dados e Inteligência Artificial podem ser combinadas para gerar valor a partir dos dados.

## Extract

Nesta etapa, os dados dos clientes são carregados para o ambiente de análise. Para fins didáticos, foi utilizada uma base contendo informações de clientes, incluindo identificação, idade e renda mensal.

```python
user_ids = [1, 2, 3, 4, 5]

print(user_ids)
```

```python
import pandas as pd

users = [
    {
        "id": 1,
        "name": "Ana",
        "idade": 25,
        "renda": 3500
    },
    {
        "id": 2,
        "name": "Carlos",
        "idade": 42,
        "renda": 8000
    },
    {
        "id": 3,
        "name": "Marina",
        "idade": 31,
        "renda": 5500
    },
    {
        "id": 4,
        "name": "Pedro",
        "idade": 58,
        "renda": 12000
    },
    {
        "id": 5,
        "name": "Lucas",
        "idade": 22,
        "renda": 2500
    }
]

df = pd.DataFrame(users)

df
```

## Transform

Nesta etapa, os dados dos clientes são enriquecidos por meio de regras de negócio e Inteligência Artificial Generativa.

Inicialmente, os clientes são classificados de acordo com sua faixa de renda. Em seguida, o modelo Gemini é utilizado para gerar recomendações financeiras personalizadas, agregando valor aos dados originais e demonstrando a aplicação prática de IA Generativa em um pipeline ETL.

```python
!pip install -q google-generativeai
```

```python
import google.generativeai as genai

genai.configure(api_key="GOOGLE_API_KEY")

for model in genai.list_models():
    print(model.name)
```

```python
model = genai.GenerativeModel("gemini-2.0-flash")

def gerar_recomendacao(cliente):

    prompt = f"""
    Você é um especialista em educação financeira.

    Cliente: {cliente['name']}
    Idade: {cliente['idade']}
    Renda: R$ {cliente['renda']}
    Faixa de renda: {cliente['faixa_renda']}

    Gere uma recomendação financeira personalizada
    em no máximo 100 caracteres.
    """

    try:
        response = model.generate_content(prompt)
        return response.text.strip()

    except Exception as e:
        return f"Erro: {str(e)}"
```

```python
df['insight_ia'] = df.apply(
    gerar_recomendacao,
    axis=1
)

df[['name', 'faixa_renda', 'insight_ia']]
```

## Load

Nesta etapa, os dados enriquecidos com recomendações geradas por IA são armazenados em um novo arquivo CSV, permitindo seu uso em análises futuras.

```python
df.to_csv(
    'clientes_com_insights.csv',
    index=False
)

print("Pipeline ETL concluído com sucesso!")
```
